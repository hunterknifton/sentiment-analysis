```python
import pandas as pd
```

```python
tweets = pd.read_csv(r'C:\Users\hknifton\Downloads\CCCB tweets v4.csv', engine='c')
tweets
```

```python
tweets_list = tweets['Hit Sentence'].tolist()
reach_list = tweets['Reach'].tolist()

for a in range(len(tweets_list)):
    tweets_list[a] = tweets_list[a].lower()
    
flagged_tweets1=[]
flagged_tweets2=[]
flagged_tweets3=[]
flagged_tweets_all=[]
other_tweets=[]

tracker1=-1
tracker2=-1
tracker3=-1
tracker4=-1
count1 = 0
reach1 = 0
count2 = 0
reach2 = 0
count3 = 0
reach3 = 0
count4 = 0
reach4 = 0
```

```python
KM1 = input("enter your first key message to check: ")
KM2 = input("enter your second key message to check: ")
KM3 = input("enter your third key message to check: ")
```

```python
# iterate through list and check if user input is in each tweet
# if it is, add 1 to counter and append the tweet to the list
# and add reach of that tweet to the total reach tracker
for i in tweets_list:
    tracker1=tracker1+1
    if KM1 in i:
        count1 = count1 + 1
        flagged_tweets1.append(i)
        reach1 = reach1 + reach_list[tracker1]

for x in tweets_list:
    tracker2=tracker2+1
    if KM2 in x:
        count2 = count2 + 1
        flagged_tweets2.append(x)
        reach2 = reach2 + reach_list[tracker2]

for y in tweets_list:
    tracker3=tracker3+1
    if KM3 in y:
        count3 = count3 + 1
        flagged_tweets3.append(y)
        reach3 = reach3 + reach_list[tracker3]
        
for z in tweets_list:
    if z not in flagged_tweets1 and z not in flagged_tweets2 and z not in flagged_tweets3:
        count4=count4+1
        other_tweets.append(z)
        reach4 = reach4 + reach_list[tracker3]

        
flagged_tweets_all = flagged_tweets1+flagged_tweets2+flagged_tweets3
unique_flagged_tweets = set(flagged_tweets_all)
unique_flagged_tweets_final = list(unique_flagged_tweets)
#unique_flagged_tweets_final
```

```python
if count1 == 0:
    avgreach1 = 0
else:
    avgreach1 = reach1/count1
    
if count2 == 0:
    avgreach2 = 0
else:
    avgreach2 = reach2/count2
    
if count3 == 0:
    avgreach3 = 0
else:
    avgreach3 = reach3/count3
    
if count4 == 0:
    avgreach4 = 0
else:
    avgreach4 = reach4/count4
```

```python
print("there were",count1,"tweets that contained your first key message, which was",KM1,"\nthe total reach was",reach1,"and the average reach was",avgreach1)
print("there were",count2,"tweets that contained your second key message, which was",KM2,"\nthe total reach was",reach2,"and the average reach was",avgreach2)
print("there were",count3,"tweets that contained your third key message, which was",KM3,"\nthe total reach was",reach3,"and the average reach was",avgreach3)
print("there were",count4,"tweets that did not contain any of your key messages","\nthe total reach was",reach4,"and the average reach was",avgreach4)
#print("here is the full list of tweets: *then add instruction to dowload counts and tweets as csv*",unique_flagged_tweets_final)
```
